```python
# Dependencies
import requests
import json
import pandas as pd
import numpy as np

from pprint import pprint

# Google developer API key
from config import gkey

# Target location
target_location = ["3600 Fairfax Dr, Arlington, VA 22201", "2401 S Smith Blvd, Arlington, VA 22202", "1250 South Hayes Street, Arlington, Virginia 22202", "S Rotary Rd, Fort Myer, VA 22211", "2001 N Sycamore St, Arlington, VA 22205", "901 S Clark St, Arlington, VA", "2100 Wilson Blvd, Arlington, VA", "Memorial Dr, Fort Myer, VA 22211", "4230 Fairfax Dr, Arlington, VA 22203", "3100 Wilson Blvd, Arlington, VA 22201", "1850 N Moore St, Arlington, VA 22209"]

# Build the endpoint URL
target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format([target_location], gkey)

# Read in Arlington County Housing Assessment data
arl_data = pd.read_csv("2018_Arl_Asmt.csv")
arl_data.head()
```

```python
# Run a request to endpoint and convert result to json
geo_data = requests.get(target_url).json()

# Pretty print the json
pprint(geo_data)
```

```python
# Print the json (using json.dumps)
print(json.dumps(geo_data, indent=4, sort_keys=True))
```

```python
# Extract latitude and longitude
lat = geo_data["results"][0]["geometry"]["location"]["lat"]
lng = geo_data["results"][0]["geometry"]["location"]["lng"]

# Print the latitude and longitude
print('''
    City: {0}
    Latitude: {1}
    Longitude: {2}
    '''.format(target_location, lat, lng))
```
